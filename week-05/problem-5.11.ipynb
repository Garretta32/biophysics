## (a) Sums of random variables

Write Python code to generate a histogram of 50,000 samples of a random variable $y = x_1 + x_2$ where $x_1$ and $x_2$ are random variables drawn from the uniform distribution over $[-1/2, 1/2]$.  Superimpose a plot of the expected probability distribution function for this variable.

```python
# Garrett Allen
import numpy as np
import matplotlib.pyplot as plt
import scipy.special as ss
from scipy.special import binom
from numpy import random
from scipy.stats import norm
import math
```

```python
samp = 500

x1 = np.random.uniform(low=-0.5, high=0.5, size=(samp,))
x2 = np.random.uniform(low=-0.5, high=0.5, size=(samp,))
y = x1 + x2
```

```python
mean = 0
sd = ((sum(y**2))/samp)**0.5
print("SD = ",sd)
bins = 50

count, bins, ignored = plt.hist(y, bins, density=True)
plt.plot(bins, 1/(sd * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mean)**2 / (2 * sd**2) ),
         linewidth=2, color='r')
plt.show()
```

## (c) Sums of nonuniform discrete random variables

rite Python code to generate a histogram of 50,000 samples of a random variables $y$ that are the sum of __2__, __4__, and __10__ random variables drawn from the discrete distribution
$$
P(1) = \dfrac{1}{3} \qquad P(2) = \dfrac{2}{9} \qquad P(3) = \dfrac{1}{9} \qquad P(4) = \dfrac{1}{3}
$$

```python
ysum = 0
samp = 500
for i in range(samp):
    xnew1 = 0
    xnew2 = 0
    xnew3 = 0
    xnew4 = 0

    x1 = np.random.uniform(low=0, high=1)
    #print(x1)
    if x1 < (1/3):
        xnew1 = 1
    x2 = np.random.uniform(low=0, high=1)
    #print(x2)
    if x2 < (2/9):
        xnew2 = 2
    x3 = np.random.uniform(low=0, high=1)
    #print(x3)
    if x3 < (1/9):
        xnew3 = 3
    x4 = np.random.uniform(low=0, high=1)
    #print(x4)
    if x4 < (1/3):
        xnew4 = 4
    #print(xnew1,xnew2,xnew3,xnew4)
    y = xnew1 + xnew2 + xnew3 + xnew4
    #print(y)
    
    ysum = np.append(ysum,[y])

print(ysum)
```

```python
mean = (1*(1/3))+(2*(2/9))+(3*(1/9))+(4*(1/3))
sd = ((sum(ysum**2))/samp)**0.5
print("SD = ",sd)
print("Mean = ",mean)
bins = 10

count, bins, ignored = plt.hist(ysum, bins, density=True)
plt.plot(bins, 1/(sd * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mean)**2 / (2 * sd**2) ),
         linewidth=2, color='r')
plt.show()
```
